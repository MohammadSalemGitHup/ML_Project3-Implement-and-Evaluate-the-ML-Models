 Preprocessing

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris, load_breast_cancer

# Standardize the features and split the data

# For Iris dataset
X_iris = df_iris.drop(columns='target')  # Features
y_iris = df_iris['target']  # Target variable

# For Breast Cancer dataset
X_cancer = df_cancer.drop(columns='target')  # Features
y_cancer = df_cancer['target']  # Target variable

# Split the data into training and test sets (80% train, 20% test)
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)
X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer = train_test_split(X_cancer, y_cancer, test_size=0.2, random_state=42)

# Standardizing the data (using StandardScaler)
scaler = StandardScaler()

#######  Normalize/Scale Features #################
# Fit and transform on training data, transform on test data
X_train_iris = scaler.fit_transform(X_train_iris)
X_test_iris = scaler.transform(X_test_iris)

X_train_cancer = scaler.fit_transform(X_train_cancer)
X_test_cancer = scaler.transform(X_test_cancer)

print("---------iris data set----------")
# Print shapes to verify
print("Iris Training set shape:", X_train_iris.shape)
print("Iris Test set shape:", X_test_iris.shape)
print("-----------cancer data set---------")
print("Breast Cancer Training set shape:", X_train_cancer.shape)
print("Breast Cancer Test set shape:", X_test_cancer.shape)
```

KNN with Different Distance Metrics (Euclidean, Manhattan, Cosine)

```python
# Import required libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score
import numpy as np

# Define the function to evaluate KNN
def evaluate_knn(X_train, X_test, y_train, y_test, metric='euclidean'):
    """
    Evaluates the K-Nearest Neighbors model with different distance metrics.

    Parameters:
    X_train, X_test: The training and testing features
    y_train, y_test: The training and testing labels
    metric: The distance metric to use ('euclidean', 'manhattan', 'cosine')

    Returns:
    classification report and ROC-AUC score
    """
    # Initialize the KNN classifier with the chosen metric
    knn = KNeighborsClassifier(metric=metric)

    # Fit the model on the training data
    knn.fit(X_train, y_train)

    # Predict on the test set
    y_pred = knn.predict(X_test)

    # Calculate classification metrics
    report = classification_report(y_test, y_pred)
    auc_score = roc_auc_score(y_test, knn.predict_proba(X_test), multi_class='ovr')

    print(f"Classification Report for {metric} distance metric:")
    print(report)
    print(f"ROC-AUC Score for {metric} distance metric: {auc_score}")

    return report, auc_score

###########################################################################
# Experiment with different distance metrics for Iris dataset
metrics = ['euclidean', 'manhattan', 'cosine']

print("\n\n\n\n\n--------------------Evaluate KNN with each distance metric----------------")
# Evaluate KNN with each distance metric
for metric in metrics:
    print(f"\nEvaluating KNN with {metric} distance metric:")
    evaluate_knn(X_train_iris, X_test_iris, y_train_iris, y_test_iris, metric)

print("\n\n\n\n\n--------------------Hyperparameter tuning to find the best value of K----------------")
# Optionally: Hyperparameter tuning to find the best value of K
param_grid = {'n_neighbors': np.arange(1, 21), 'metric': ['euclidean', 'manhattan', 'cosine']}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, n_jobs=-1)

# Fit the model on the training data to find the best K and distance metric
grid_search.fit(X_train_iris, y_train_iris)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"\nBest parameters from GridSearchCV: {best_params}")
print(f"Best cross-validation score: {best_score}")

# Evaluate the best model found using the test data
best_knn = grid_search.best_estimator_
y_pred_best = best_knn.predict(X_test_iris)


print("\n\n\n\n\n--------------------report----------------------")

# Print the final classification report for the best model
final_report = classification_report(y_test_iris, y_pred_best)
final_auc_score = roc_auc_score(y_test_iris, best_knn.predict_proba(X_test_iris), multi_class='ovr')

print(f"\nFinal Classification Report for best KNN model:")
print(final_report)
print(f"Final ROC-AUC Score for best KNN model: {final_auc_score}")
```

 Support Vector Machines (SVM)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score

print("\n\n ----------------------Implement SVM----------------------")
from sklearn.svm import SVC
from sklearn.metrics import classification_report, roc_auc_score

# Initialize SVM models with probability=True
svm_linear = SVC(kernel='linear', probability=True)
svm_poly = SVC(kernel='poly', probability=True)
svm_rbf = SVC(kernel='rbf', probability=True)

# Train the SVM models
svm_linear.fit(X_train_iris, y_train_iris)
svm_poly.fit(X_train_iris, y_train_iris)
svm_rbf.fit(X_train_iris, y_train_iris)

# Predict and evaluate
y_pred_linear = svm_linear.predict(X_test_iris)
y_pred_poly = svm_poly.predict(X_test_iris)
y_pred_rbf = svm_rbf.predict(X_test_iris)

# Print classification reports
print("Linear SVM Classification Report:")
print(classification_report(y_test_iris, y_pred_linear))

print("Polynomial SVM Classification Report:")
print(classification_report(y_test_iris, y_pred_poly))

print("RBF SVM Classification Report:")
print(classification_report(y_test_iris, y_pred_rbf))

# ROC-AUC scores using predict_proba
roc_auc_linear = roc_auc_score(y_test_iris, svm_linear.predict_proba(X_test_iris), multi_class='ovr')
roc_auc_poly = roc_auc_score(y_test_iris, svm_poly.predict_proba(X_test_iris), multi_class='ovr')
roc_auc_rbf = roc_auc_score(y_test_iris, svm_rbf.predict_proba(X_test_iris), multi_class='ovr')

print(f"ROC-AUC (Linear SVM): {roc_auc_linear}")
print(f"ROC-AUC (Polynomial SVM): {roc_auc_poly}")
print(f"ROC-AUC (RBF SVM): {roc_auc_rbf}")


```

 Ensemble Methods

Boosting using AdaBoost

```python
# Import required libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
import numpy as np

# Initialize AdaBoost with a base estimator (Decision Tree)
adaboost_clf = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),  # Weak learner: Decision Stump
    n_estimators=50,  # Number of boosting rounds
    random_state=42
)

# Train the AdaBoost model
adaboost_clf.fit(X_train_iris, y_train_iris)

# Predict on the test set
y_pred_adaboost = adaboost_clf.predict(X_test_iris)

# Evaluate AdaBoost
print("------ AdaBoost (Boosting) Results ------")
print("Classification Report for AdaBoost:")
print(classification_report(y_test_iris, y_pred_adaboost))

# Accuracy
accuracy_adaboost = accuracy_score(y_test_iris, y_pred_adaboost)
print(f"Accuracy: {accuracy_adaboost:.4f}")

# ROC-AUC (Multiclass)
try:
    y_prob_adaboost = adaboost_clf.predict_proba(X_test_iris)  # Predict probabilities
    roc_auc_adaboost = roc_auc_score(y_test_iris, y_prob_adaboost, multi_class='ovr')
    print(f"ROC-AUC: {roc_auc_adaboost:.4f}")
except AttributeError:
    print("ROC-AUC score cannot be calculated due to missing probability estimates.")
```

 Random Forest (Bagging)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# Initialize Random Forest with desired hyperparameters
random_forest_clf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)

# Train the model
random_forest_clf.fit(X_train_iris, y_train_iris)

# Predict on the test set
y_pred_rf = random_forest_clf.predict(X_test_iris)

# Classification report
print("Random Forest Classification Report:")
print(classification_report(y_test_iris, y_pred_rf))

# ROC-AUC score
roc_auc_rf = roc_auc_score(y_test_iris, random_forest_clf.predict_proba(X_test_iris), multi_class='ovr')
print(f"ROC-AUC (Random Forest): {roc_auc_rf}")
```
